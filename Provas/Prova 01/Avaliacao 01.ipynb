## Questão 01

```python
# Importação das bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataframe = pd.read_csv('wine_ap1.csv')
dataframe = dataframe.iloc[:,1:]
```

```python
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
y = dataframe['y']
dataframe[y == 0] = dataframe[y == 0].fillna(dataframe[y == 0].mean())
dataframe[y == 1] = dataframe[y == 1].fillna(dataframe[y == 1].mean())
dataframe[y == 2] = dataframe[y == 2].fillna(dataframe[y == 2].mean())
dataframe
```

```python
# Mostrar as correlações, dizer quais colunas estavam mais correlacionadas e as que estavam menos

X = dataframe.drop('y', axis=1)
y = dataframe['y']

import seaborn as sns
sns.heatmap(dataframe.corr())
```

```python
# Padronização dos dados

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
x_transformed = scaler.transform(X)
x_transformed = pd.DataFrame(x_transformed, index = range(0, len(x_transformed)), columns = X.columns)
x_transformed.head()
```

## Questão 02

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

X_train_transformed, X_test_transformed, y_train_transformed, y_test_transformed = train_test_split(x_transformed, y, test_size=0.25, random_state=42, stratify=y)
```

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
y_pred_train = dtc.predict(X_train)
```

```python
dtc.fit(X_train_transformed, y_train_transformed)
y_pred_transformed = dtc.predict(X_test_transformed)
y_pred_train_transformed = dtc.predict(X_train_transformed)
```

```python
print(metrics.classification_report(y_pred, y_test))
```

```python
print(metrics.classification_report(y_pred_transformed, y_test_transformed))
```

```python
def accuracy(y_real, y_pred):
    count = 0
    for i in range(0, len(y_real)):
        if np.array(y_real)[i] == np.array(y_pred)[i]:
            count += 1
    
    return count / len(y_pred_train)
```

```python
accuracy(y_test, y_pred)
```

```python
accuracy(y_test_transformed, y_pred_transformed)
```

```python
accuracy(y_train, y_pred_train)
```

```python
accuracy(y_train_transformed, y_pred_train_transformed)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred_train_k = knn.predict(X_train)
```

```python
knn.fit(X_train_transformed, y_train_transformed)
y_pred_train_transformed_k = knn.predict(X_train_transformed)
```

```python
knn.fit(X_train, y_train.values.ravel())
y_pred_test_k = knn.predict(X_test)
```

```python
knn.fit(X_train, y_train.values.ravel())
y_pred_test_transformed_k = knn.predict(X_test_transformed)
```

```python
accuracy(y_train, y_pred_train_k)
```

```python
accuracy(y_train_transformed, y_pred_train_transformed_k)
```

```python
accuracy(y_test, y_pred_test_k)
```

```python
accuracy(y_test_transformed, y_pred_test_transformed_k)
```
